# Анализ клиентов банка и их сегментация по потреблению <a class="tocSkip"> 

###### Описание проекта. <a class="tocSkip"> 

Проанализируйте клиентов регионального банка и сегментировать пользователей по количеству потребляемых продуктов.
    
В нашем распоряжении датасет bank_dataset.csv, который содержит данные о клиентах банка «Метанпром». Банк располагается в Ярославле и областных городах: Ростов Великий и Рыбинск.

Заказчик исследования - менеджер продукта    
 
###### Цель исследования : <a class="tocSkip"> 
Минимизация оттока клиентов банка на основе анализа их поведения, которая включает в себя:
    -исследование факторов, усиливающих отток клиентов и удерживающих клиентов в банке
    -проверка гипотез:
    
          - о различии дохода между клиентами, которые пользуются двумя продуктами банка , и теми, которые пользуются одним
          - о различии дохода  денежных средств на счетах между клиентами филиалов банка в разных городах
    
    ПРИМЕЧАНИЕ: возможно, в процессе исследовательского анализа возникнет новая гипотеза, которой будет заменена вторая 


###### План исследования : <a class="tocSkip"> 

   ***1. Обзор данных:***
    
    - общий обзор датасета (размер, содержание)
    - проверка  на соответствие типов данных
    - проверка на полноту данных (наличие/отсутствие пропусков)

    
   ***2. Предобработка данных:***
    
    - приведение наименований столбцов к единому стилю
    - анализ столбца с пропусками для принятия решения их заполнения/удаления
    - проверка на дубликаты
    
    
   ***3. Исследовательский анализ данных:***

    - распределение признаков в разрезе оттока в целях выявления тенденций оттока/ удержания
    - исследование коррелирующих признаков, построение тепловой карты
    - построение диаграмм рассеяния : визуализация зависимости между разными признаками (ПЕРЕЧЕНЬ БУДЕТ ОПРЕДЕЛЕН В ПРОЦЕССЕ. Возможно будет интересным увидеть зависимость дохода от прочих признаков)
    - распределение признаков в разрезе филиалов банка
    - распределение признаков в разрезе данных по потреблению
    
   ***4. Проверка гипотез***
    
    - проверка гипотезы о различии дохода между клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним
    - проверка гипотезы о различии дохода  денежных средств на счетах между клиентами филиалов банка в разных городах.
    
    Общий алгоритм тестирования:
    - делим данные на группы согласно гипотезе
    - строим точечный график дохода/остатков клиента в группах, смотрим выбросы
    - подготовка данных к проведению теста (проверяем наличие и удаляем выбросы данных межквартильным методом (IQR-методом))
    - определение метода проведения теста
    - проведение тестирования
    - интерпретация результатов
    
   ***5. Кластеризация***
    
    - подготовка данных (переводим категориальные переменные в числовые), стандартизация данных
    - кластеризация клиентов с целью определения факторов оттока
    - определение признаков оттока/надежности клиентов.
