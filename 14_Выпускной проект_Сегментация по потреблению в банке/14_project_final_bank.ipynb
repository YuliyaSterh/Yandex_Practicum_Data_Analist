Для удобства зададим несколько функций:

```python
#функция для вывода состава столбцов
def describe_cols(df_data):
    df = df_data
    values = pd.DataFrame()
    
    for col in df.columns:
        values[col] = [df[col].unique(), df[col].dtypes, df[col].nunique()]
        
    return values.transpose().sort_values(by=1, ascending=True).rename(
        {0: 'Значения', 1: 'тип данных', 2: 'Кол-во уникальных значений'}, axis=1)


#функция для вывода пустых значений и их доли в столбцах датафрейма
def nan_values(df_data):
    df = df_data
    count_missing = df.isna().sum()
    percent_missing = round(df.isna().sum() * 100 / len(df), 2)
    
    missing_value_df = pd.DataFrame({'column_name': df.columns,
                                     'num_missing': count_missing,
                                     'percent_missing': percent_missing})
    
    return missing_value_df


#функция для вывода информации о файле
def df_info (df_data):
    print('Количество строк файла:', df_data.shape[0])
    print('Количество столбцов файла:', df_data.shape[1])
    print()
    print('Информация о содержимом файла:')
    display(describe_cols(df_data))
    print()
    print('Информация о пропусках данных:')
    display(nan_values(df_data))
    print()
    print('Первые пять строк таблицы:')
    display(df_data.head())
```

Таким образом:

- пропуски не связаны с признаком оттока, т.е. присутствуют и у ушедших, и у действующих клиентов;
- у клиентов с признаком оттока существуют остатки на счетах;
- нет зависимости наличия пропуска от какого-либо другого признака;
- пропуски данных есть только в записях о клиентах из Рыбинска и Ярославля.

Скорее всего причина в следующем: как правило, менеджеры (а мы говорим о небольшом, судя по базе, региональном банке, локация которого находится в небольших городах) ведут свои базы в excel, но остатки на балансовых счетах меняются ежедневно, поэтому актуальные остатки нужно выгрузить из баланса банка дополнительно и добавить их к базе. 

В таком случае существуют два порядка формирования наших данных:

- к существующей таблице менеджер подгружает отстатки. В этом случае если остатка нет, поле таблицы менеджена останется незаполненным, т.е. мы получаем пропуск

- к выгрузке остатков на счетах клиентов присоединяется таблица менеджера. При этом мы видим, что в наших данных нет нулевых остатков на счетах, то есть были выгружены только ненулевые остатки, к которым присоединены данные менеджера, и , возможно, какая-то часть данных была утеряна.

Поскольку пропуски мы видим только в данных Рыбинска и Ярославля, а  в данных Ростова они отсутствуют, полученная таблица составлена из трех баз, в которой  Рыбинск и Ярославль пошли по первому пути, а Ростов -по второму.

В реальном исследовании для корректности анализа стоит связаться с заказчиком и обговорить алгоритм формирования данных для анализа, которому будут придерживаться все подразделения. Но в нашем случае это невозможно, поэтому мы предположим,  что пропуски поля `balance`- это нулевые остатки. Нулевые остатки есть на балансе любого банка, но у Ростова они отсутствуют (не загружены), поэтому во избежание искажения анализа мы не будем удалять строки с пропусками (т.к.это 36% датасета), и не будем заменять пропуски на нули (т.к.это исказит данные поля balance в разрезе городов).

### Общее распределение признаков клиентов банка

```python
#функция для формирования графиков распределения:
#data - датафрейм, на основе которого строится распределение
#hue- признак, по которому мы хотим увидеть детализацию данных

def tables(data, hue=None):
    #дискретные значения
    for column in ['city','gender', 'credit_card', 'loyalty', 'objects', 'products', 'churn']:
        ax = sns.countplot(data=data, x=column, hue=hue);
        plt.title(f'Распределение клиентов по признаку {column}', fontsize = 16, color = 'black')
        plt.show()
    
    plt.figure(figsize=(16,7))
    sns.countplot(data=data, x='age', hue=hue);
    plt.title('Распределение клиентов по признаку age', fontsize = 16, color = 'black')
    plt.show()

    #непрерывные значения
    for column in ['score',  'balance', 'estimated_salary']:
        sns.kdeplot(data=data, x=column, hue=hue);
        plt.title(f'Распределение клиентов по признаку {column}', fontsize = 16, color = 'black')
        plt.show()
```

## Проверка гипотез

### Проверка гипотезы о различии дохода между клиентами, которые пользуются двумя продуктами банка,   и теми, которые пользуются одним продуктом


**Сформулируем гипотезы:**

`Н0` :статистически значимых различий в доходе клиентов, которые пользуются одним продуктом и теми, кто пользуется двумя продуктами, нет 

`Н1` : статистически значимые различия в доходе клиентов, которые пользуются одним продуктом и теми, кто пользуется двумя продуктам, есть

Подготовим и проанализируем данные:

```python
#формируем выборки дохода клиентов, пользующихся 1 и 2 продуктами
salary_1 = df.query('products == 1')['estimated_salary']
salary_2 = df.query('products == 2')['estimated_salary']
```

```python
salary_1.describe()
```

```python
salary_2.describe()
```

Поскольку набор данных большой, критическим уровнем статистической значимости примем значение 0,001.

```python
print(f'Уровень статистической значимости: t-test {st.ttest_ind(salary_1, salary_2)[1]}')
```

p-value значительно больше принятого уровня критической значимости (0.001). Нулевая гипотеза не отвергнута, статистически значимых различий в доходе  пользователей с большой долей вероятности  нет

## Кластеризация

Поскольку нашей задачей является сегментация пользователей по потреблению, попробуем для ее решения применить кластеризацию и проанализируем полученный результат.

### Построение матрицы расстояний и построение дендрограммы

Для начала построим матрицу расстояний и нарисуем дендрограмму.



Подготовим датафрейм: категоризируем колонку `city`, удалим колонку `user_id`, а также колонку `balance`, в которой остались пропуски (пропуски были не у всех филиалов, и велика вероятность, что они связаны с разными алгоритмами формирования данных. Во избежание искажений данных при формировании кластеров не будем использовать эти данные)

```python
#сохраняем в отдельную переменную подготовленный датафрейм
df_clast = df.drop(['user_id', 'balance'], axis=1)
```

```python
#категоризируем колонку `city`:
print('Категории городов до изменений:', '\n', df_clast['city'].value_counts())

df_clast.loc[df_clast['city']=='Ярославль','city'] = 1
df_clast.loc[df_clast['city']=='Рыбинск','city'] = 2
df_clast.loc[df_clast['city']=='Ростов Великий','city'] = 3
#меняем тип данных на int
df_clast['city'] = df_clast['city'].astype(int)

print('Категории городов после изменений:', '\n', df_clast['city'].value_counts())
```

```python
df_clast.info()
```

Стандартизируем данные:

```python
sc = StandardScaler()
x_sc = sc.fit_transform(df_clast)
```

Строим матрицу расстояний на стандартизованной матрице признаков:


```python
linked = linkage(x_sc, method = 'ward') 
```

Рисуем дендрограмму:

```python
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top',truncate_mode="level", p=5) 
plt.title('Hierarchial clustering for GYM')
plt.show() 
```

**Вывод**

Примем за количество кластеров n_clusters=3

### Кластеризация

```python
# задаём модель k_means с числом кластеров 3 и фиксируем значение random_state
km = KMeans(n_clusters=3, random_state=0)
# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 2)
labels = km.fit_predict(x_sc)
# сохраняем метки кластера в поле нашего датасета
df_clast['cluster_km'] = labels

# выводим статистику по средним значениям наших признаков по кластеру
display(df_clast.groupby('cluster_km').mean().T.reset_index())
```
